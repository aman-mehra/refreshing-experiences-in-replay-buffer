```python
# UPDATE BASE_PATH and OUT_BASE as per user needs
BASE_PATH = "D:/Aman/IIIT Delhi/IP-RL/GridWorldExperiments/Logs/"
OUT_BASE = "D:/Aman/IIIT Delhi/IP-RL/GridWorldExperiments/"

action_map = {0:"R",1:"D",2:"L",3:"U"}
experiment_lbl_map = {"Uniform":"U","Refresh":"R","Refresh-Noisy":"N","Small-Big":"S","CER":"C","CER-Refresh":"CR","Geometric-Decay":"G","PER":"P"}
experiment_linestyle_map = {"Uniform":"-","Refresh":"--","Refresh-Noisy":"-.","Small-Big":":","CER":":","CER-Refresh":"-.","Geometric-Decay":"-.","PER":"-."}
# linecolor_map = {"1":"-","Refresh":":","Small-Big":"--"}
```

```python
"""
Helper class to visualize experiment logs
"""
class LogViewer:
    """
    Function to compare results of multiple experiments in a single graph
    
    PARAMETERS
    experiment - Is the name of the experiment situateed in the topmost subdirectory of the log folder.
                 Existing set of experiments include the keys of the dictionary 'experiment_lbl_map' in the previous cell
    algo       - Identifies which type of reward was used. Options are 'TD' and 'tree'. This forms the second level subdirectory in the logs folder
    n          - Number of steps used in reward computation. Third level subfolder in logs.
    buff_sz    - Size of relay buffer used in experiment. Name of log file is buff_sz.pickle
    no_noise   - Filtering parameter to remove any experiments with the Noisy key word in their name. 
    smoothing  - Averages values over a window of this size to create smoother graphs.
    
    SPECIAL KEYWORDS
    across     - This is keyword provided to the first four parameters to automatically pick a predefined set of experiments to compare.
                 This set can be seen in the function compare_logs under the comment 'Across Set'
    """
    def compare_logs(self,experiment="Uniform",algo='TD',n=1,buff_sz='across',no_noise=True,smoothing=10):
        buff_szs = [buff_sz]
        algos = [algo]
        steps = [n]    
        experiments = [experiment]
        
        if isinstance(n, list):
            steps = n
        if isinstance(experiment, list):
            experiments = experiment
        if isinstance(algo, list):
            algos = algo
        if isinstance(buff_sz, list):
            buff_szs = buff_sz      
        
        # Across set 
        if buff_sz == 'across':
            buff_szs = [100,1000,10000]
        if n=='across':
            steps=[1,2,3,4]
        if algo=='across':
            algos = ['TD','tree']
        if experiment == 'across':
            experiments = ["Uniform","Refresh","Small-Big","Refresh-Noisy"]
            if no_noise:
                experiments = ["Uniform","Refresh","Small-Big"]
            
        self.make_comparative_plot(experiments,algos,steps,buff_szs,smoothing)
     
    
    """
    Given a set of experiments, this function generates and saves plots.
    """
    def make_comparative_plot(self,experiments,algos,steps,buff_szs,smoothing):
        
        fig1, ax1 = plt.subplots()
        fig2, ax2 = plt.subplots()
        fig3, ax3 = plt.subplots()
        fig4, ax4 = plt.subplots()
        
        ax1.set_title("Epsiode Returns")
        ax1.set_xlabel("Episodes")
        ax1.set_ylabel("Returns")
        
        ax2.set_title("Epsiode Variance")
        ax2.set_xlabel("Episodes")
        ax2.set_ylabel("Variance")
        
        ax3.set_title("Epsiode Entropy")
        ax3.set_xlabel("Episodes")
        ax3.set_ylabel("Entropy")
        
        ax4.set_title("Epsiode Running Variance")
        ax4.set_xlabel("Episodes")
        ax4.set_ylabel("Running Variance")
        
        for exp in experiments:
            for algo in algos:
                for n in steps:
                    for buff_sz in buff_szs:
                        file_path = BASE_PATH + exp + '/' + algo + '/' + str(n) + '/' + str(buff_sz) + '.pickle'
                        plt_lbl = experiment_lbl_map[exp] + '-' + algo + '-' + str(n) + '-' + str(buff_sz)
                        logs = {}
                        if os.path.exists(file_path):
                            with open(file_path,'rb') as f:
                                logs = pickle.load(f)
                            ax1.plot(self.smooth_array(logs["returns"],smoothing),label=plt_lbl,linestyle=experiment_linestyle_map[exp])
                            ax2.plot(self.smooth_array(logs["variance"],smoothing),label=plt_lbl,linestyle=experiment_linestyle_map[exp])
                            ax3.plot(self.smooth_array(logs["entropy"],smoothing),label=plt_lbl,linestyle=experiment_linestyle_map[exp])
                            ax4.plot(self.smooth_array(logs["running_variance"],smoothing),label=plt_lbl,linestyle=experiment_linestyle_map[exp])
                    
        # Shrink current axis by 20%
        box = ax1.get_position()
        ax1.set_position([box.x0, box.y0, box.width * 0.8, box.height])
        box = ax4.get_position()
        ax4.set_position([box.x0, box.y0, box.width * 0.8, box.height])
                    
        ax1.legend(loc='upper left',bbox_to_anchor=(1, 1))
        ax2.legend(loc='upper right')
        ax3.legend(loc='lower right')
        ax4.legend(loc='upper left',bbox_to_anchor=(1, 1))
        
        fig1.savefig(OUT_BASE+"log_output/1.png",dpi=600)
        fig2.savefig(OUT_BASE+"log_output/2.png",dpi=600)
        fig3.savefig(OUT_BASE+"log_output/3.png",dpi=600)
        fig4.savefig(OUT_BASE+"log_output/4.png",dpi=600)
                    
        plt.show()
    
    
    """
    Generates plot for a single experiment. Refer to Parameter defn for compare_logs function above. Note that the 'across' keyword 
    is not compatible with this function
    """
    def view_logs(self,experiment="Uniform",algo='TD',n=1,buff_sz=100,smoothing=10):
        file_path = BASE_PATH + experiment+ '/' + algo + '/' + str(n) + '/' + str(buff_sz) + '.pickle'
        logs = {}
        with open(file_path,'rb') as f:
            logs = pickle.load(f)
        self.make_plot(logs,smoothing)
    
    def smooth_array(self,arr,smoothing):
        out = []
        for i in range(1,len(arr)):
            smooth_lim = max(0,i-smoothing)
            out.append(sum(arr[smooth_lim:i+1])/len(arr[smooth_lim:i+1]))
        return out
    
    def make_plot(self,logs,smoothing=10):
        fig1, ax1 = plt.subplots()
        ax1.set_title("Train Log - Epsiode Returns")
        ax1.plot(self.smooth_array(logs["returns"],smoothing))
        ax1.set_xlabel("Episodes")
        ax1.set_ylabel("Returns")
        plt.show()
        
        fig2, ax2 = plt.subplots()
        ax2.set_title("Train Log - Epsiode Variance")
        ax2.plot(self.smooth_array(logs["variance"],smoothing))
        ax2.set_xlabel("Episodes")
        ax2.set_ylabel("Variance")
        plt.show()
        
        fig3, ax3 = plt.subplots()
        ax3.set_title("Train Log - Epsiode Entropy")
        ax3.plot(self.smooth_array(logs["entropy"],smoothing))
        ax3.set_xlabel("Episodes")
        ax3.set_ylabel("Entropy")
        plt.show()
        
        fig4, ax4 = plt.subplots()
        ax4.set_title("Train Log - Epsiode Running Variance")
        ax4.plot(self.smooth_array(logs["running_variance"],smoothing))
        ax4.set_xlabel("Episodes")
        ax4.set_ylabel("Running Variance")
        plt.show()
        
```

```python
"""
Function to generate a video of how the contents of the replay buffer change over the course of the experiment.
"""
def generate_video(experiment,method,n,buf_sz,fps=20):
    
    image_folder = BASE_PATH + experiment + "/" + method +"/" + str(n) + "/" + str(buf_sz)+"-memory/"
    video_path = BASE_PATH + experiment + "/" + method +"/" + str(n) + "/" + str(buf_sz)+".avi"
    log_path = BASE_PATH + experiment + "/" + method +"/" + str(n) + "/" + str(buf_sz) + ".pickle"

    images = [img for img in os.listdir(image_folder)]
    frame = cv2.imread(os.path.join(image_folder, images[0]))
    height, width, layers = frame.shape

    logs = {}
    with open(log_path,'rb') as f:
        logs = pickle.load(f)
    fig, ax = plt.subplots()
    
    video = cv2.VideoWriter(video_path, 0, fps, (width*3,height))
    fig, ax = plt.subplots()
#     fig2, ax2 = plt.subplots()
    
    img_arr = None
    for image in tqdm(images):
        ep_num = int(image.split("_")[-1].split(".")[0])
        returns = logs["returns"][:ep_num]
        running_variance = logs["running_variance"][:ep_num]
        
        ax.plot(running_variance)
        ax.set_title("Running Variance")
        ax.set_ylabel("Running Variance")
        ax.set_xlabel("Episodes")
        
        fig.canvas.draw()
        
        data2 = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
        data2 = data2.reshape((height,fig.canvas.get_width_height()[0]) + (3,))
        ax.clear()
        
        ax.plot(returns)
        ax.set_title("Returns")
        ax.set_ylabel("Returns")
        ax.set_xlabel("Episodes")
                
        fig.canvas.draw()

        data = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
        data = data.reshape((height,fig.canvas.get_width_height()[0]) + (3,))
        ax.clear()

        img_arr = cv2.imread(os.path.join(image_folder, image))
        combined = np.hstack((img_arr,data,data2))
        video.write(combined)      
        
    plt.close(fig)

    cv2.destroyAllWindows()
    video.release()
```

```python
generate_video("CER","TD",3,100,fps=70)
```

```python
def view_policy(algo='TD',n=1,buff_sz=100):
    file_path = BASE_PATH + algo + '/' + str(n) + '/' + str(buff_sz) + '-q.pickle'
    q  = None
    with open(file_path,'rb') as f:
        q = pickle.load(f)
    policy = [["" for i in range(q.shape[0])] for j in range(q.shape[1])]
    for i in range(q.shape[0]):
        for j in range(q.shape[1]):
            action = np.argmax(q[i,j,:])
            policy[i][j] = action_map[action]
            
    for i in range(q.shape[0]):
        for j in range(q.shape[1]):
            if i == q.shape[0]//2 and j == q.shape[1]//2:
                print("X",end=" ")
            else:
                print(policy[i][j],end=" ")
        print()
```
